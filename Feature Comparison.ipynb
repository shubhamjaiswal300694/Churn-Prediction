```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/HEINZ/Desktop/model_input.csv')
```

```python
df.head()
```

```python
tenure_churned = df[df['churn_label']==1]['TENURE_IN_MONTHS']
tenure_non_churned = df[df['churn_label']==0]['TENURE_IN_MONTHS']

sns.set(rc={'figure.figsize':(20,10)})
sns.distplot(tenure_churned, color='Red', label='Churned')
sns.distplot(tenure_non_churned, color='Green', label='Not-churned')

plt.legend()
plt.show()
```

```python
gl_bal_2_churned = np.array(df[df['churn_label']==1]['GL_BAL_S_2'].dropna())
gl_bal_2_not_churned = np.array(df[df['churn_label']==0]['GL_BAL_S_2'].dropna())

gl_bal_2_churned = gl_bal_2_churned - min(gl_bal_2_churned)
gl_bal_2_not_churned = gl_bal_2_not_churned - min(gl_bal_2_not_churned)
gl_bal_2_churned = np.log(gl_bal_2_churned + 1)
gl_bal_2_not_churned = np.log(gl_bal_2_not_churned + 1)


sns.set(rc={'figure.figsize':(20,10)})
sns.distplot(gl_bal_2_churned, color='Red', label='Churned')
sns.distplot(gl_bal_2_not_churned, color='Green', label='Not-churned')

plt.legend()
plt.show()
```
